# F . Retrieve the top 3 movies which has got more revenue in 2014,2015,2016

```python
df14=df[df['Year'] == 2014].nlargest(3,'Revenue (Millions)')
df14
```

```python
df15=df[df['Year'] == 2015].nlargest(3,'Revenue (Millions)')
df15
```

```python
df16=df[df['Year'] == 2016].nlargest(3,'Revenue (Millions)')
df16
```

```python
top_3_by_each_year  = pd.concat([df14,df15,df16]) # merge top 3 movie in year 2014,15,16
top_3_by_each_year 
```

# most top 3 movie list according to the Revenue in all three year 2014 2015 2016 

```python
filtered_movies = df[df['Year'].isin([2014, 2015, 2016])]
top_3 = top_3_by_each_year.sort_values(by='Revenue (Millions)', ascending=False)
top_3.head(3) 
```

# G. Top three movies acted by Will Smith and rating > 7
