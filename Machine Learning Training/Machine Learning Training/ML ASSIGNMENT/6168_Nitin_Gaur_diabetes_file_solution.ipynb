```python
# Z-Score (data_point - mean)/standard_deviation
print(df['col_9'])
```

```python
mean = np.mean(df['col_9'])
std = np.std(df['col_9'])
print("mean ",mean)
print("std ",std)
```

```python
df.describe() # we can match that mean value for col_9 is same for both case describe() and also when we calculate seperatly 
```

## Z-Score for each each data point of col_9

```python
# Z-Score (data_point - mean)/standard_deviation
#for 0--------
z_scr_0 = (0-mean)/std
print("Z_score col_9 for  data point = 0 ",z_scr_0)
```

```python
#for 0--------
z_scr_1 = (1-mean)/std
print("Z_score col_9 for  data point = 1 ",z_scr_0)
```

```python
plt.plot(z_scr_1,z_scr_0)
plt.show()
```
