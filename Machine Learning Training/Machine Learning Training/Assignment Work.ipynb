#  Prompt the user to input any numbers between 1 and 9. If User enters any other number, prompt 
# the user to input again. If the user enters the valid number, display the below two patterns.


```python
      
while True:
    n = int(input ( 'please enter the value in range of 1 to 9-> '))
    if n>9 or n<1:
        print("PLease enter valid number in range of 1 to 9")
    else:
        print_pattern_two(n)
        print(end='\n')
        print_pattern_one(n)
    break
    
def print_pattern_one(n):
    for i in range(n+1):
        for j in range(1,i+1):
            print(i,end=' ')  
        print()   
def print_pattern_two(n):
    for i in range(n+1):
        for j in range(1,i+1):
            print(j,end=' ')  
        print()    
```

# 5. Write a Python program that accepts a word from the user and reverse it

```python
def word_reverse(word):
    word = word[::-1]
    print("here is the reverse word->   ",word)
    
while True:
    word = str(input('enter the word = '))
    if word=='':
        print("I need one word for reversing so please enter a word => ")
    else:
        word_reverse(word)
    break
    
```
